# Task
Analyze the dataset from "https://raw.githubusercontent.com/deepamukil7-bit/Deepa/main/fashion_items.csv" to find the most popular categories for each customer type.

## Load the data

### Subtask:
Load the dataset from the given URL into a pandas DataFrame.


**Reasoning**:
Import pandas and load the dataset from the given URL into a DataFrame.



```python
import pandas as pd

url = "https://raw.githubusercontent.com/deepamukil7-bit/Deepa/main/fashion_items.csv"
df = pd.read_csv(url)
df.head()
```

## Explore the data

### Subtask:
Display the first few rows and the column names to understand the structure of the data.


**Reasoning**:
Display the first few rows and the column names of the DataFrame to understand its structure.



```python
display(df.head())
display(df.columns)
```

## Clean the data

### Subtask:
Handle any missing values or inconsistencies in the 'Category' and 'Customer_Type' columns.


**Reasoning**:
Check for missing values in 'Category' and 'Target Customer' columns and display their unique values to identify inconsistencies.



```python
print("Missing values in 'Category':", df['Category'].isnull().sum())
print("Missing values in 'Target Customer':", df['Target Customer'].isnull().sum())
print("\nUnique values in 'Category':", df['Category'].unique())
print("Unique values in 'Target Customer':", df['Target Customer'].unique())
```

## Analyze the data

### Subtask:
Group the data by 'Customer_Type' and then by 'Category' to count the occurrences of each category for each customer type.


**Reasoning**:
Group the DataFrame by 'Target Customer' and 'Category' and count the occurrences of each combination to find the most popular categories for each customer type.



```python
category_counts = df.groupby(['Target Customer', 'Category']).size().reset_index(name='Count')
display(category_counts)
```

## Find the most popular categories

### Subtask:
For each customer type, identify the category with the highest count.


**Reasoning**:
For each target customer, find the category with the highest count and store the results.



```python
most_popular_categories = category_counts.loc[category_counts.groupby('Target Customer')['Count'].idxmax()]
display(most_popular_categories)
```

## Present the results

### Subtask:
Display the most popular category for each customer type.


## Analyze Average Price by Customer Type

### Subtask:
Calculate the average price of items for each 'Target Customer' type.

**Reasoning**:
Group the DataFrame by 'Target Customer' and calculate the mean of the 'Price' column to find the average price for each customer type.

```python
average_price_by_customer = df.groupby('Target Customer')['Price'].mean().reset_index()
display(average_price_by_customer)
```

## Analyze Color and Brand Preference

### Subtask:
Analyze if there are any trends in color or brand preference for each 'Target Customer' type.

**Reasoning**:
Group the DataFrame by 'Target Customer' and 'Color' to count color preferences, and then by 'Target Customer' and 'Brand' to count brand preferences. Display the top colors and brands for each customer type.

```python
color_counts = df.groupby(['Target Customer', 'Color']).size().reset_index(name='Count')
brand_counts = df.groupby(['Target Customer', 'Brand']).size().reset_index(name='Count')

print("Color Preference by Customer Type:")
display(color_counts.loc[color_counts.groupby('Target Customer')['Count'].idxmax()])

print("\nBrand Preference by Customer Type:")
display(brand_counts.loc[brand_counts.groupby('Target Customer')['Count'].idxmax()])
```

**Reasoning**:
Display a clear and concise summary of the most popular category for each target customer type based on the `most_popular_categories` DataFrame.



```python
print("Most Popular Categories for Each Customer Type:")
for index, row in most_popular_categories.iterrows():
    customer_type = row['Target Customer']
    category = row['Category']
    count = row['Count']
    print(f"- For '{customer_type}', the most popular category is '{category}' with a count of {count}.")
```
